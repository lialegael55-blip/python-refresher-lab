```python
# Variables and data types
x = 10
y = 5
print(x + y)

# Boolean logic
is_valid = True
print(is_valid)

# Conditional statements
if x > y:
    print("x is greater than y")
```

```python
# Lists
numbers = [1, 2, 3, 4]
print(numbers)

# Loop
for num in numbers:
    print(num)

# Range
for i in range(5):
    print(i)
```

```python
# Write to file
with open("file.txt", "w") as f:
    f.write("Hello Python")

# Read file
with open("file.txt", "r") as f:
    print(f.read())
```

```python
import sqlite3

conn = sqlite3.connect("test.db")
cursor = conn.cursor()

cursor.execute("CREATE TABLE IF NOT EXISTS users (name TEXT)")
cursor.execute("INSERT INTO users VALUES ('Gael')")

conn.commit()

for row in cursor.execute("SELECT * FROM users"):
    print(row)

conn.close()
```
